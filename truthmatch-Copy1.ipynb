```python
import ROOT,glob,math,os
from array import array
```

```python
path="./signal/"
paths=glob.glob(path+"*root")
print(paths)
f=ROOT.TFile(paths[1],"r")
branches=[]
for i in f.GetListOfKeys():
    print(i.GetName())
    branches.append(i.GetName())
t=f.Get("BPHY13")
N=t.GetEntries()
hx=ROOT.TH1F("x_mass","x_mass",200,0,20)
for i in range(N):
    t.GetEntry(i)
    truth_tq_mass=t.truth_mass[0]/1000.0
    hx.Fill(truth_tq_mass)
c1=ROOT.TCanvas("c1","c1",100,0,900,700)
c1.SetLeftMargin(0.12)
c1.SetLogy()
ROOT.gStyle.SetOptStat(0)
hx.Draw()
c1.Draw()
```
